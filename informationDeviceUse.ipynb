```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothic')
```

```python
informationDeviceUse = pd.read_csv('/content/drive/MyDrive/likelionC/seniorActivities/informationDeviceUse.csv', encoding='UTF-8')
informationDeviceUse
```

```python
subtotal = informationDeviceUse[['유형별(1)', '2020']]
subtotal
```

```python
# 0~3행 자르기
subtotal.drop(subtotal.index[0:3], inplace=True)
subtotal
```

```python
# 문자받기, 문자보내기를 useMessage로 묶음
useMessage = subtotal.loc[3:4, ['유형별(1)', '2020']]

# float로 변경 후, 합
useMessage['2020'] = useMessage['2020'].astype(float)
total = useMessage['2020'].sum()
useMessage.loc[2] = ['문자사용능력', total]

# 인덱스 정렬(0부터)
useMessage.index = range(len(useMessage))

# 0~1행 자르기
useMessage.drop(useMessage.index[0:2], inplace=True)
useMessage
```

```python
# 인덱스 정렬(0부터)
subtotal.index = range(len(subtotal))

# 문자받기, 문자보내기 지우기
subtotal.drop(subtotal.index[0:2], inplace=True)
subtotal
```

```python
# useMessage + subtotal
informationDeviceUse = useMessage.append(subtotal, ignore_index = True)
informationDeviceUse
```

```python
df = informationDeviceUse.copy()
df.dtypes
```

```python
#컬럼 '2020' dtype을 object에서 float로 변경
df['2020'] = df['2020'].astype(float)
df.dtypes
```

```python
df.sort_values('2020', ascending = False, inplace=True)
# 원 그래프를 그릴 데이터 추출
sizes = df['2020']
labels = df['유형별(1)']
explode = [0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.05, 0.12, 0.19, 0.26]

# seaborn의 컬러 팔레트 가져오기
colors = sns.color_palette('hls', len(labels))

# 원 그래프
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode, shadow=True, counterclock=False, startangle=90, pctdistance=0.8)  #선택인자

# 제목 추가
plt.title('노인의 정보화 기기 사용 역량', position=(1, 1), fontsize=17)

# 그래프 표시
plt.show()
```
