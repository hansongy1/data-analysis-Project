## 오프라인 대비 온라인 비율

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumBarunGothic')
```

```python
df = pd.read_csv('/content/drive/MyDrive/likelionC/seniorActivities/getInformation.csv', encoding='cp949')
df
```

```python
# '2021', '2022' 열 제거
df.drop(['2021', '2022'], inplace=True, axis='columns')

dfAge = df.query("구분별1 == '연령'")
dfAge.index = range(len(dfAge))
df2 = dfAge.copy()
df2
```

## 전체 활동량

```python
new_df = df2.drop('구분별1', axis=1).copy()

# 온라인, 오프라인 구분(5, 6 포함, 포함하지 않음으로 / 5, 6 : 오프라인)
onLine = new_df.drop(columns=new_df.columns[new_df.columns.str.contains('202\d\.(5|6)')])
offLine = new_df.filter(regex='구분|202\d\.(5|6)')

# '구분별2' 열 제거하고 copy
onLine = onLine.drop('구분별2', axis=1).copy()
offLine = offLine.drop('구분별2', axis=1).copy()

# float로 변경
onLine = onLine.astype(float)
offLine = offLine.astype(float)

# 전체 활동량
# onLineSum, offLineSum 하나로 합치기 - concat
total = pd.concat([onLine, offLine], axis=1, join= 'inner')

# 전체활동량 합계
total['합계'] = total.iloc[:, 0:16].astype(float).sum(axis=1)
total
```

```python
# 온라인 활동량의 합
onLine['합계'] = onLine.iloc[:, 0:12].astype(float).sum(axis=1)
onLine
```

```python
# 오프라인 대비 온라인 비율
df_total = total.loc[:, ['합계']]
df_onLine = onLine.loc[:, ['합계']]
offtoOn = (df_onLine / df_total) * 100  # 온라인비율(%) = (온라인 활동량 / 전체 활동량) * 100

# df2에 있던 '구분별2' 열 가져오기
offtoOn['구분별2'] = df2['구분별2']

dfFinal = offtoOn[['구분별2', '합계']]
dfFinal
```

```python
dfFinal.sort_values('합계', ascending = False, inplace=True)

# 원 그래프를 그릴 데이터 추출
sizes = dfFinal['합계']
labels = dfFinal['구분별2']
explode = [0.05, 0.05, 0.05, 0.05, 0.09, 0.14, 0.20]

# seaborn의 컬러 팔레트 가져오기
colors = sns.color_palette('Blues_r', 10)

# 원 그래프(변경 후)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, explode=explode, shadow=True, counterclock=False, startangle=90, pctdistance=0.8)  #선택인자

# 제목 추가
plt.title('2021-2022 오프라인 대비 온라인 \n 정보습득 비율', position=(1.0, 1), fontsize=16)

# 그래프 표시
plt.show()
```
